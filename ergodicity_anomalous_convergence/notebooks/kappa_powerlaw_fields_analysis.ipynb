```r
df_pl_kappa <- readRDS("../datasets/ising1DrateErgodicityPowerlawsTimeFields.rds")
```

```r
get_power_law <- function(df_pl_kappa, N, dynamics) {
    df <- df_pl_kappa
    df <- df[df["N"]==N,]
    df <- df[df["dynamics"]== dynamics,]
    y <- as.vector(df["alpha_mean"])$alpha_mean
    y_upper <- as.vector(df["alpha_upper"])$alpha_upper
    y_lower <- as.vector(df["alpha_lower"])$alpha_lower
    x <- as.vector(df["H"])$H
    list("x"=x, "y"=y, "y_upper"=y_upper, "y_lower"=y_lower)
}
```

```r
plot_exponents <- function(fname, dynamic, base_plot="../plots/fieldsPowerLaws/") {
    fname <- paste(base_plot, fname, sep = "")
    pdf(fname, pointsize = 20) # pointsize will effect all label and title font sizes.
    val <- get_power_law(df_pl_kappa, 512, dynamic)
    plot(val$x, val$y, lty=1, pch=1, type="p", col="black", ann=FALSE, ylim=c(0.0, 2.2), xlim=c(0.4, 1.5))
    arrows(val$x, val$y_lower, val$x, val$y_upper, col="black",length=0.05, angle=90, code=3)
    val <- get_power_law(df_pl_kappa, 1024, dynamic)
    lines(val$x, val$y, lty=2, pch=2, type="p", col="blue",ann=FALSE, ylim=c(0.0, 2.2), xlim=c(0.4, 1.5))
    arrows(val$x, val$y_lower, val$x, val$y_upper, col="blue", length=0.05, angle=90, code=3)
    val <- get_power_law(df_pl_kappa, 1536, dynamic)
    lines(val$x, val$y, lty=3, pch=3, type="p", col="red", ann=FALSE, ylim=c(0.0, 2.2), xlim=c(0.4, 1.5))
    arrows(val$x, val$y_lower, val$x, val$y_upper, col="red", length=0.05, angle=90, code=3)
    xlabel <- "Fields Range"
    ylabel <-  bquote(paste("Scaling exponent ", alpha))
    title_main <- bquote(paste("Inverse rate power laws ", Kappa(t), " ~ ", Ct^{alpha}))
    subtitle <- bquote(paste(
        "Average Magnetisation Ergodicity ", .(dynamic)
    ))
    title(title_main, xlab=xlabel, ylab=ylabel)
    mtext(subtitle)
      legend("topleft",
        legend = c("N=512", "N=1024", "N=1536"),
        pch = c(1, 2, 3),
        col = c("black", "blue", "red"),
        lwd = 2,
        bty = "n",
      )
    dev.off()
}
```
